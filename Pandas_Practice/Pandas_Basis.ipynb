# Transformaciones y manipulacion de datos con pandas

```python
# para la creacion de una nueva columna puedo utiulizar la notacion df['nombre_nueva_columna']

import pandas as pd

path = "online_retail.csv"

df = pd.read_csv(path)
df
```

```python
df['Precio total'] = df['Quantity'] * df['UnitPrice']
```

```python
# en esta parte se puede ver como se ha agregado al final la nueva columna requerida
df
```

```python
# Para optener informacion relacionada con los datos tipo de dato nombre de columnas puedo utilizar la funcion .info()
df.info()
```

```python
# De lo anterior se puede ver que la columna InvoiceDate tiene un tipo de dato incorrecto
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
print(df.info())
```

```python
# tambien puedo aplicar una funcion lambda a cada item de una columna especifica usando la funcion apply()

df['Precio con descuento'] = df['UnitPrice'].apply(lambda x: x*0.9)
df
```

```python
# Con la funcion apply tambien puedo aplicar una fucnion a filas con valores numericos u categoricosabs

# en este ejemplo voy a aplicar una funcion para categorizar una variable cuantitativaabs
def categizarPrecio(precio):
    if precio > 0 and precio <= 10:
        return 'Bajo'
    elif precio > 10 and precio <= 25:
        return 'Medio'
    else:
        return 'Alto'


df['Categoria precio'] = df['UnitPrice'].apply(categizarPrecio)
df
```

# Analisis de ventas con pandas: agrupaciones y estadistica

```python
# para la creacion de una nueva columna puedo utiulizar la notacion df['nombre_nueva_columna']

import pandas as pd

path = "online_retail.csv"

df = pd.read_csv(path)
df
```

```python
country_count = df['Country'].value_counts()
country_count
```

```python
# unso de la funcion groupby()
# lo que hacer este codigo es hacer un agrupamiento de los diferentes paises en country y luego tomar la columna de quatity y sumar
#donde haya una coincidencia
ctry_group = df.groupby('Country')['UnitPrice'].sum()
ctry_group
```

```python
# lo que se hace en este codigo es con groupby() se agrupa todo por los valores de country y luego se toma la columna
# [UnitPrice]a la cual le agregamos con agg() dos fucnones mas como mean y sum

countr_stats = df.groupby('Country')['UnitPrice'].agg(['mean','sum'])
countr_stats
```

```python
country_stock = df.groupby(['Country', 'StockCode'])['UnitPrice'].agg(['sum'])
country_stock
```

```python
country_stock2 = df.groupby(['Country', 'StockCode'])[['UnitPrice', 'Quantity']].agg(['sum'])
country_stock2
```

```python
# Tambien puedo aplicar funciones a grupaciones 

def total_revenue(group):
    return (group['Quantity'] * group['UnitPrice']).sum()


rev_per_country = df.groupby(['Country']).apply(total_revenue)
print(rev_per_country)
```
