# Seleccion de datos en Pandas, iloc y loc

```python
# Busqueda con iloc, esta funcion me permite buscar filas y columnas por indice, es un metodo de el dataframe

fst_five_rows = df.iloc[0:5]
fst_five_rows
```

```python
# En pandas la convencion simpre sera [fila, columna]
# en este caso puedo seleccionar un numero especifico de filas y columnas
iloc_rows = df.iloc[0:5, 0:2]
iloc_rows
```

```python
# Tambien puedo utilizar la funcion del data frame loc para buscar pero en este caso por etiqueta de columna
# in this case you could notice that the select rows tale the total number of index marked
fst_5_rows = df.loc[0:5, ['InvoiceNo', 'StockCode']]
fst_5_rows
```

# Manejo de datos faltantes

```python
# La funcion isna() is not available pregunta si algun dato no exite en el dataframe o serie, 
# toma como nulo los valores de None, np.nan, Nat(dato faltante de tiempo)

missing_d_sales = d_sales.isna()
missing_d_sales
```

```python
missing_d_sales.sum()
```

```python
missing_d_df = df.isna()
missing_d_df
```

```python
# La funcion sum me ayuda a encontrar cuantos datos faltantes tengo en total
missing_d_df.sum()
```

```python
# eliminar filas del dataframe, puedo utilizar la funcion dropna, esta por defecto eliminal las filas de un Df especifico

no_mnissing_values = df.dropna()
no_mnissing_values
```

```python
# En esta seccion confirmo que los datos faltantes en las filas fueron eliminados
no_mnissing_values.isna().sum()
```

```python
# Adicionalmente dropna() me permite eliminar columnas haciendo uso de su parametro axis
# df.dropna(axis=1) -> los valores que puesde tomar son 0 -> indica fila y 1 -> indica columna

missing_data = df
missing_data.isna().sum()
```

```python
# al consultar de nuevo si hay algun dato como nulo o nota available podemos ver como se ha eliminado las columnas completas
missing_data.dropna(axis=1).isna().sum()
```

```python
# puedo hacer uso de la funcion fillna() el cuento toma un valor numerico, str o tambien una lista para llenar los laroles nulos en 

miss_dta = df
miss_dta.isna().sum()
```

```python
# En este ejemplo puedo ver como la funcion fillna lleno la informacion faltante en el df
miss_dta.fillna(0).isna().sum()
```

```python
# puedo tambien llenar columnas especificas

# first of all take the required colum

colum = df['Description']
colum
```

```python
# I check if the selected colum has missing info haciendo uso de isna() junto a sum()
colum.isna().sum()
```

```python
# Hago uso de fillna() para llenar los campos faltantes con el valor de null, pero arreglo en el df general
# Para poder modificar el df directaemnte es importante agregar el parametro inplace=True el cual indica el cambio en los valores donde no hay datos
df['Description'].fillna('null', inplace=True)
```

```python
df.isna().sum()
```

```python
# ahora eliminare las filas donde el customer id no esta disponible
#df.dropna(inplace=True)
# tambien puedo utlizar el parametro subset = [colum1, colum2] donde paso una lista de columnas a revisar
df.dropna(subset = ['CustomerID'], inplace = True)
```

```python
df.isna().sum()
```

# Transformaciones y manipulacion de datos con pandas

```python
# para la creacion de una nueva columna puedo utiulizar la notacion df['nombre_nueva_columna']

import pandas as pd

path = "online_retail.csv"

df = pd.read_csv(path)
df
```

```python
df['Precio total'] = df['Quantity'] * df['UnitPrice']
```

```python
# en esta parte se puede ver como se ha agregado al final la nueva columna requerida
df
```

```python
# Para optener informacion relacionada con los datos tipo de dato nombre de columnas puedo utilizar la funcion .info()
df.info()
```

```python
# De lo anterior se puede ver que la columna InvoiceDate tiene un tipo de dato incorrecto
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
print(df.info())
```
