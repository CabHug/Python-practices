# Fusion de DataFrames en pandas: merge, concat y join

```python
# La union de varios data frames me permite trabajar con informacion mas consolidada 
```

```python
import pandas as pd

df1 = pd.DataFrame({
    'key': ['A', 'B', 'C'],
    'Value': [1,2,4]
})

df2 = pd.DataFrame({
    'key': ['C', 'D', 'E'],
    'Value': [6,2,7]
})

print(df1)
print(df2)
```

```python
"""
El método merge en Pandas se utiliza para combinar dos DataFrames en función de una o más claves, similar a cómo se haría en una base de datos. Permite realizar diferentes tipos de uniones:

Inner Join: devuelve solo las filas con claves comunes en ambos DataFrames.
Outer Join: devuelve todas las filas de ambos DataFrames, llenando con NaN donde no hay coincidencias.
Left Join: devuelve todas las filas del primer DataFrame y las coincidentes del segundo.
Right Join: devuelve todas las filas del segundo DataFrame y las coincidentes del primero.
Esto es esencial para integrar información de diversas fuentes y realizar análisis más complejos.
"""

# Inner Join: devuelve solo las filas con claves comunes en ambos DataFrames.
inner_merge = pd.merge(df1, df2, on = 'key', how = 'inner')
inner_merge
```

```python
# Outer Join: devuelve todas las filas de ambos DataFrames, llenando con NaN donde no hay coincidencias.
outer_merge = pd.merge(df1,df2,on='key',how='outer')
outer_merge
```

```python
# Left Join: devuelve todas las filas del primer DataFrame y las coincidentes del segundo.
left_merge = pd.merge(df1,df2, on='key', how='left')
left_merge
```

```python
# Right Join: devuelve todas las filas del segundo DataFrame y las coincidentes del primero.
right_merge = pd.merge(df1,df2, on='key', how='right')
right_merge
```

```python
# Concepto de concatenacion
"""
La función concat en Pandas se utiliza para concatenar dos o más DataFrames a lo largo de un eje (vertical u horizontal). Sus parámetros principales son:

objs: Una lista de DataFrames a concatenar.
axis: El eje a lo largo del cual concatenar (0 para filas, 1 para columnas).
join: Especifica cómo manejar índices no coincidentes (default es 'outer').
ignore_index: Si es True, se ignorarán los índices originales y se generarán nuevos índices.
Esta función es útil para combinar datasets y facilitar el análisis.
"""
df3 = pd.DataFrame({
    'A': ['A0', 'A1', 'A2'],
    'B': ['B0', 'B1', 'B2']
})

df4 = pd.DataFrame({
    'A': ['A0', 'A4', 'A5'],
    'B': ['B3', 'B4', 'B5']
})

print(df3)
print(df4)
```

```python
vertical_concat = pd.concat([df3,df4])
vertical_concat
```

```python
horizontal_concat = pd.concat([df3,df4], axis=1)
horizontal_concat
```

```python
"""
El método join en Pandas combina DataFrames usando índices. Este método utiliza diferentes parámetros, siendo los más comunes:

other: el DataFrame a combinar.
how: el tipo de unión ('inner', 'outer', 'left', 'right').
on: especifica la clave o índice para la unión.
lsuffix y rsuffix: sufijos para columnas con el mismo nombre.
Este método es útil para fusionar datos basándose en índices, permitiendo una integración eficiente de información.
"""
```
