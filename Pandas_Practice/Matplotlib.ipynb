```python
import numpy as np
import matplotlib.pyplot as plt

month = np.array(['E', 'F','M','A','Ma','J','Jl'])
sales = np.array([23,45,67,87,56,4,23])

# configurar el tamaño
plt.figure(figsize=(8,6))

# creo la grafica
plt.plot(month, sales, marker = 'o', color = 'blue')

#Cambio el titulo
plt.title('Ventas mensuales de un producto')

# cambio el titulo o label del eje x
plt.xlabel('mes')

# cambio el titulo o label del eje y
plt.ylabel('sales')
```

```python
# Ahora voy a imprimir un grafico de dispersion

import matplotlib.pyplot as plt

horas=[1,2,3,4,5,6,7,8]
resultado=[50,55,60,65,70,75,80,85]

plt.scatter(horas, resultado, marker='*')
#Cambio el titulo
plt.title('resultado vs horas de estudio')

# cambio el titulo o label del eje x
plt.xlabel('horas')

# cambio el titulo o label del eje y
plt.ylabel('resultado')
```

```python
# personalizacion de graficos
import matplotlib.pyplot as plt

horas=[1,2,3,4,5,6,7,8]
resultado_E1=[50,55,60,65,70,75,80,85]
resultado_E2=[40,53,65,70,78,87,90,94]

plt.scatter(horas, resultado_E1, marker='*', color='blue', label= 'Estudiante 1')
plt.scatter(horas, resultado_E2, marker='o', color='green', label= 'Estudiante 2')
#Cambio el titulo
plt.title('resultado vs horas de estudio')

# cambio el titulo o label del eje x
plt.xlabel('horas')

# cambio el titulo o label del eje y
plt.ylabel('resultado')
```

```python
# Grafico de barras
import matplotlib.pyplot as plt

plt.figure(figsize=(8,6))

categorias = ['A', 'B', 'C']
sales = [120, 150, 140]

plt.bar(categorias, sales, label='Ventas mensuales')

plt.title('ventas por productos')
plt.xlabel('Productos')
plt.ylabel('Ventas')
plt.legend()
plt.show
```

```python
plt.barh(categorias, sales, label='Ventas mensuales')

plt.title('ventas por productos')
plt.ylabel('Productos')
plt.xlabel('Ventas')
plt.legend()
plt.show
```

```python
# definicion de un diagrama de pastel
plt.pie(sales, labels=categorias, autopct='%1.1f%%', startangle=140)
plt.title('Grafico de pastel')
# asegurar que el grafico sea un circulo
plt.axis('equal')
plt.show()
```

```python
# Graficos de histogramas y boxplot para distribuciones

import numpy as np
import matplotlib.pyplot as plt

# otra forma de hacer un numero random reproducible es poniendo la semilla
np.random.seed(400)
data = np.random.normal(170, 10,200)
data
```

```python
plt.hist(data, color='salmon', bins=10, edgecolor='black', alpha=1)
plt.title('Grafico histograma')
plt.xlabel('altuta')
plt.ylabel('Densidad')
plt.show()
```

```python
np.random.seed(30)

edades = [np.random.normal(30,5,100),
          np.random.normal(40,5,100),
          np.random.normal(35,5,100)
         ]
edades
```

```python
plt.boxplot(edades, patch_artist=True, notch=True, vert=True,
           label=['A', 'B', 'C'])

plt.title('Grafico bocplot')
plt.xlabel('grupo')
plt.ylabel('edades')
plt.show()
```
